# 任意の関数に対する1次近似と接線群

この notebook では、関数 `f(x)` とその導関数 `df(x)` を自由に変更できます。

授業では、
- 売上関数（二次関数）
- 効用関数（対数関数）
- 指数関数
- 任意の多項式

などに差し替えて利用できます。


````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib_fontja
# =====================================================
# ここだけ変更すれば任意の関数を分析できる
# =====================================================

def f(x):
    return x*(100-x)

def df(x):
    return 100-2*x

# 描画範囲
x_min = 0
x_max = 100

# =====================================================
````

````python
x_grid = np.linspace(x_min, x_max, 1000)

def plot_tangent_reconstruction(points,
                                tangent_width=10.4,
                                pointsize=10,
                                show_true_function=True,
                                title=""):
    
    fig, ax = plt.subplots(figsize=(9,5))

    if show_true_function:
        ax.plot(x_grid, f(x_grid),
                linewidth=0.1,
                label='f(x)')

    for x0 in points:
        y0 = f(x0)
        slope = df(x0)
        xs = np.linspace(x0-tangent_width,
                         x0+tangent_width,
                         100)
        ys = y0 + slope*(xs-x0)
        ax.plot(xs, ys, linewidth=2)
        ax.scatter([x0],[y0], s=pointsize)

    ax.set_title(title)
    ax.grid(True, alpha=0.3)
    #ax.legend()
    plt.show()
````

## 粗い観測

接線の本数が少ない状態


````python
points = np.linspace(x_min, x_max, 4)

plot_tangent_reconstruction(
    points,
    tangent_width=10.6,
    title='粗い一次近似'
)
````

## 中程度の観測

````python
points = np.linspace(x_min, x_max, 10)

plot_tangent_reconstruction(
    points,
    tangent_width=10.4,
    title='細かい一次近似'
)
````

## 高密度観測

接線が増えると関数形状が浮かび上がる


````python
points = np.linspace(x_min, x_max, 80)

plot_tangent_reconstruction(
    points,
    tangent_width=4,
    title='かなり細かい一次近似'
)
````

## 関数を隠す

接線情報だけから形状を推測する


````python
points = np.linspace(x_min, x_max, 40)

plot_tangent_reconstruction(
    points,
    tangent_width=0.15,
    show_true_function=False,
    title='Only local linear information'
)
````
